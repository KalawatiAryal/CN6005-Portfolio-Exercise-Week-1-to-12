```python
#Apriori and Association Rules
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


# 1) Transactions (N = 5)
transactions = [
    ["Mango", "Onion", "Nintendo", "Key-chain", "Eggs", "Yo-yo"],              # T1
    ["Doll", "Onion", "Nintendo", "Key-chain", "Eggs", "Yo-yo"],               # T2
    ["Mango", "Apple", "Key-chain", "Eggs"],                                   # T3
    ["Mango", "Umbrella", "Corn", "Key-chain", "Yo-yo"],                       # T4
    ["Corn", "Onion", "Key-chain", "Ice-cream", "Eggs"]                        # T5
]

N = len(transactions)
min_sup = 0.60        # 60% = 3/5
min_conf = 0.80       # strong rules threshold


# 2) One-hot encode transactions
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)


# 3) Frequent itemsets (Apriori)
freq = apriori(df, min_support=min_sup, use_colnames=True)

# Add count column (support * N)
freq["count"] = (freq["support"] * N).round().astype(int)

# Sort for nice display
freq_sorted = freq.sort_values(by=["itemsets", "support"], ascending=[True, False]).reset_index(drop=True)

print("=== FREQUENT ITEMSETS (support >= 60%) ===")
print(freq_sorted)


# 4) Association rules
rules = association_rules(freq, metric="confidence", min_threshold=min_conf)

# Keep common columns & sort
rules = rules[["antecedents", "consequents", "support", "confidence", "lift"]]
rules = rules.sort_values(by=["confidence", "lift", "support"], ascending=False).reset_index(drop=True)

print("\n=== STRONG RULES (confidence >= 80%) ===")
print(rules)


# 5) Optional: make the output easier to read
def fs(x):  # format set
    return "{" + ", ".join(sorted(list(x))) + "}"

pretty_rules = rules.copy()
pretty_rules["antecedents"] = pretty_rules["antecedents"].apply(fs)
pretty_rules["consequents"] = pretty_rules["consequents"].apply(fs)

print("\n=== STRONG RULES (Pretty) ===")
print(pretty_rules.to_string(index=False))
```
